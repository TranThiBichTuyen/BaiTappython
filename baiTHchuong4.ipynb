```python
#bai1
from math import sqrt
print("Chương trình tính diện tích Tam Giác")
a=float(input("Nhập cạnh a>0:"))
b=float(input("Nhập cạnh b>0:"))
c=float(input("Nhập cạnh c>0:"))
if (a<=0 or b <=0 or c <=0) or (a+b)<=c or (a+c)<=b or b+c<=a:
 print("Tam giác không hợp lệ")
else:
 cv=a+b+c
 p=cv/2
 dt=sqrt(p*(p-a)*(p-b)*(p-c))
 print("Diện tích =",dt)
```

```python
#bai3
def BMI(height,weight):
  return weight/(height**2)
def PhanLoai(bmi):
  if bmi<18.5:
     return "Gầy"
  elif bmi<=24.9:
    return "Bình thường"
  elif bmi<=29.9:
    return "Hơi Béo"
  elif bmi<=34.9:
   return "Béo Phì Cấp Độ 1"
  elif bmi<=39.9:
   return "Béo Phì Cấp Độ 2"
  else:
   return "Béo Phì Cấp độ 3"
def NguyCoBenh(bmi):
 if bmi<18.5:
  return "Thấp"
 elif bmi<=24.9:
  return "Trung Bình"
 elif bmi<=29.9:
  return "Cao"
 elif bmi<=34.9:
  return "Cao"
 elif bmi<=39.9:
  return "Rất cao"
 else:
  return "Nguy Hiểm"
print("Nhập vào chiều cao:")
height=float(input())
print("Nhập vào cân nặng:")
weight=float(input())
bmi=BMI(height,weight)
print("BMI của bạn=",bmi)
print("Phân loại bạn=",PhanLoai(bmi))
print("Nguy cơ bệnh của Thím=",NguyCoBenh(bmi))
```

```python
#bai4
def ROI(dt,cp):
 return (dt-cp)/cp
def GoiYDauTu(roi):
 if roi>=0.75:
  return "Nên đầu tư"
 else:
  return "Không nên đầu tư"
print("Chương trình tính ROI")
dt=int(input("Nhập Doanh Thu:"))
cp=int(input("Nhập chi phí:"))
roi=ROI(dt,cp)
print("Tỉ Lệ ROI=",roi)
print("==>",GoiYDauTu(roi))
```

```python
#bai5
def fibonacci(n):
 if n<=2 :
  return 1
 return fibonacci(n-1)+fibonacci(n-2)
def listfibo(n):
 for i in range(1,n+1):
  print(fibonacci(i),end='\t')
print(fibonacci(9))
listfibo(9)
```

```python
#bai7
from math import sqrt

print("=== TÍNH ĐỘ DÀI ĐOẠN THẲNG AB ===")
xA = float(input("Nhập hoành độ xA: "))
yA = float(input("Nhập tung độ yA: "))
xB = float(input("Nhập hoành độ xB: "))
yB = float(input("Nhập tung độ yB: "))

AB = sqrt((xB - xA)**2 + (yB - yA)**2)
print(f"Độ dài đoạn AB = {AB:.2f}")

```

```python
#bài 8
from math import log

print("=== CHƯƠNG TRÌNH TÍNH logₐ(x) ===")

a = float(input("Nhập cơ số a > 0 và a ≠ 1: "))
x = float(input("Nhập giá trị x > 0: "))

if a <= 0 or a == 1 or x <= 0:
    print("Giá trị không hợp lệ! (Cần a > 0, a ≠ 1 và x > 0)")
else:
    kq = log(x) / log(a)
    print(f"log_{a}({x}) = {kq:.4f}")
```

```python
#cau 10
import time
import os

def clear():
    # Xóa màn hình (Windows)
    os.system('cls')

# 4 hình mẫu (bạn có thể thay bằng hình khác theo đề)
hinh1 = """
*
**
***
****
*****
"""

hinh2 = """
    *
   ***
  *****
 *******
*********
"""

hinh3 = """
*****
 ****
  ***
   **
    *
"""

hinh4 = """
*   *
 * *
  *
 * *
*   *
"""

# Danh sách 4 hình
ds_hinh = [hinh1, hinh2, hinh3, hinh4]

print("=== CHƯƠNG TRÌNH VẼ HÌNH DÙNG SLEEP ===")
for i, hinh in enumerate(ds_hinh, start=1):
    clear()
    print(f"Hình {i}:\n{hinh}")
    time.sleep(5)  # dừng 5 giây trước khi hiện hình tiếp theo

clear()
print("Kết thúc chương trình!")
```

```python
#câu 12
def oscillate(a, b):
    # Tăng từ a đến b-1
    for i in range(a, b):
        yield i
    # Giảm từ b-1 ngược về a, rồi đi thêm 2 bước nữa
    for i in range(b-1, a-5, -1):
        yield i

# --- Chạy thử ---
for n in oscillate(-3, 5):
    print(n, end=' ')
```

```python
#cau 13
def tong_uoc(n):
    s = 0
    for i in range(1, n):
        if n % i == 0:
            s += i
    return s

def kiem_tra_so(n):
    tong = tong_uoc(n)
    if tong == n:
        print(f"{n} là số hoàn thiện")
    elif tong > n:
        print(f"{n} là số thịnh vượng")
    else:
        print(f"{n} là số thiếu (Deficient number)")

# --- Chạy thử ---
n = int(input("Nhập số nguyên dương n: "))
kiem_tra_so(n)
```

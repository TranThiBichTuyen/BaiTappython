Câu 6: Trình bày ý nghĩa toán tử /, //, %, **, and, or, is.

***1. Toán tử số học (trong Python)***
  / (Phép chia thông thường)
: Trả về kết quả chia số thực, ngay cả khi phép chia cho ra số nguyên.

**Ví dụ**: 10 / 2 sẽ cho kết quả là 5.0.
// (Phép chia lấy phần nguyên)
: Trả về phần nguyên của phép chia.

***Ví dụ:*** 10 // 3 sẽ cho kết quả là 3.
% (Phép chia lấy số dư)
: Trả về phần dư của phép chia.

***Ví dụ***: 10 % 3 sẽ cho kết quả là 1.
** (Phép lũy thừa)
: Nâng một số lên lũy thừa của một số khác, tương đương với việc nhân lặp.

***Ví dụ:*** 2 ** 3 (2 mũ 3) sẽ cho kết quả là 8.

***2. Toán tử logic (trong Python)***
and (Logic VÀ)

+ Trả về True nếu cả hai toán hạng đều là True; ngược lại, trả về False.
Ví dụ: True and False sẽ cho kết quả là False.
or (Logic HOẶC)

+ Trả về True nếu ít nhất một trong hai toán hạng là True; chỉ trả về False nếu cả hai toán hạng đều là False.
Ví dụ: True or False sẽ cho kết quả là True.

***3. Toán tử so sánh (trong Python)***
is (So sánh định danh)
+ Kiểm tra xem hai biến có trỏ đến cùng một đối tượng trong bộ nhớ hay không.
Ví dụ: Nếu a = b và a và b cùng trỏ đến một đối tượng, thì a is b sẽ trả về True.

. Các Loại Lỗi Lập Trình Phổ Biến
Lỗi cú pháp (Syntax Error)

+ Mô tả: Xảy ra khi bạn viết sai quy tắc cú pháp của ngôn ngữ Python, chẳng hạn như viết thiếu dấu ngoặc, sai cách đặt biến, hoặc sai cấu trúc lệnh.
Ví dụ: Lỗi SyntaxError, IndentationError (lỗi thụt đầu dòng sai cách).
Lỗi thời gian chạy (Runtime Error - Ngoại lệ)

+ Mô tả: Lỗi này xuất hiện trong quá trình chương trình thực thi. Khi gặp lỗi, Python sẽ tạo ra một đối tượng ngoại lệ (exception object).
Ví dụ: IndexError (truy cập phần tử của danh sách sai chỉ mục), TypeError (thực hiện phép toán trên hai kiểu dữ liệu không tương thích), NameError (sử dụng biến chưa được khai báo).
Lỗi logic (Logic Error)

+ Mô tả: Lỗi này là sai sót trong thuật toán hoặc logic của chương trình, dẫn đến kết quả không đúng như mong đợi, nhưng chương trình vẫn chạy mà không báo lỗi.
Ví dụ: Tính toán sai công thức, đưa ra điều kiện sai trong câu lệnh if.

```python
# Gán giá trị ban đầu cho các biến
x = 10
y = 5
number_of_closed_cases = 20
ncc = 5

print(f"Giá trị ban đầu: x = {x}, y = {y}, number_of_closed_cases = {number_of_closed_cases}, ncc = {ncc}\n")

# Các lệnh đã được viết ngắn gọn
x += 1
print(f"(a) x sau khi x += 1: {x}")

x /= 2
print(f"(b) x sau khi x /= 2: {x}")

x -= 1
print(f"(c) x sau khi x -= 1: {x}")

x += y
print(f"(d) x sau khi x += y: {x}")

x -= (y + 7)
print(f"(e) x sau khi x -= (y + 7): {x}")

x *= 2
print(f"(f) x sau khi x *= 2: {x}")

number_of_closed_cases += 2 * ncc
print(f"(g) number_of_closed_cases sau khi += 2 * ncc: {number_of_closed_cases}")
```
